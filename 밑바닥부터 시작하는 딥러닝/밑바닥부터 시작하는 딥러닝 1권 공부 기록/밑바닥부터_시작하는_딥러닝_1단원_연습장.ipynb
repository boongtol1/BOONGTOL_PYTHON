```python
# coding: utf-8
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = "--", label="cos") # cos 함수는 점선으로 그리기
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title('sin & cos')
plt.legend()        # plt.plot 의 label 요소와 관련되어 있다!
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

# 'fig 1-5.png' 이미지 읽어오기
img = imread('fig 1-5.png')

# NumPy 배열의 일부 출력
print(img[0:9, 0:9, 0:9])  # 이미지 배열의 처음 5x5 부분을 출력

# 이미지 출력
plt.imshow(img)
plt.show()
```
